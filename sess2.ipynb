Q14) The Root mean square value of the Linear regression model is
 a) 18529629.51147644
 b) 1001531.331097424
 c) 13524820.12533588
 d) 14599645.26554688
 
 Ans: A

```python

from statsmodels.tools.eval_measures import rmse

import statsmodels.api as sm
from sklearn.model_selection import train_test_split

X=data['Area']
X=sm.add_constant(X)
Y=data['Production']
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.3,random_state=42)
SLR_model=sm.OLS(Ytrain,Xtrain).fit()
rmse(Ytest,SLR_model.predict(Xtest))
```

Q15) The MAE of the Linear regression model is
 a) 18529629.51147644
 b) 1121707.916712476
 c) 13524820.12533588
 d) 112599645.2655468
 Ans: B

```python
from sklearn import metrics
from sklearn.linear_model import LinearRegression
lnr=LinearRegression()
lnr.fit(Xtrain,Ytrain)
Ypred=lnr.predict(Xtest)
print(metrics.mean_absolute_error(Ytest,Ypred))
```
